```python
def calc(a,b,p) :
    if(p == " + ") :
      return( a + b)
    elif(p ==  "-"):
       print( a - b )
    elif(p ==  "*" ):
       print( a * b)
    elif(p == "/" ):
       print ( a /  b)
    return "done"
n1 = int(input(" Enter the first no.:" ))
opt = (input(" Enter the operation you want :"))
n2 = int(input(" Enter the second no.:"))

calc(n1,n2,opt)
```
