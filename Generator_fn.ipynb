```python
def feb(term1,term2,terms) :
  d = []
  for i in d :
    term3 = term1 + term2
    d.append(term3)
    print("terms are :",d)
```

```python
feb(0,1,3)
```

```python
def febo(n) :
  a,b =  0,1
  for i in range(n) :
    print(a)
    a,b = b , a+b
```

```python
febo(9)
```

```python
for i in febo(9) :
 print(i)
```

```python
def test_fi():
    a,b = 0,1
    while True :
       yield a
       a,b = b , a+b
```

```python
fib  = test_fi()
```

```python
print(fib)
```

```python
for i in test_fi(4) :
   print(next(fib))
```

```python
def count_test(n) :
  count = 1
  while count <= n :
    print(count)
    count = count + 1
```

```python
for i in count_test(3) :
  print(i)
```
